# Bar Graph

```python
#plotting a bar graph
fig,ax=plt.subplots(2,2,figsize=(30,32),sharey=True)
ax[0,0].bar(x["country"],x["2023 population"],width=0.75,color='brown',label='2023 Population',edgecolor='black')
ax[0,0].set_xlabel("Countries",fontsize=17,fontname='Arial',fontweight='bold')
ax[0,0].set_ylabel("2023 Population (In Billions)",fontsize=17,fontname='Arial',fontweight='bold')
ax[0,0].set_xticks(x["country"])
ax[0,0].set_xticklabels(x["country"],rotation=90,fontsize=15)
ax[0,0].legend(fontsize=17)
ax[0,0].grid(True)
ax[0,0].set_title("Population in 2023",fontweight='bold',fontsize=17)

ax[0,1].bar(x["country"],x["2000 population"],width=0.75,color='goldenrod',label='2000 Population',edgecolor='black')
ax[0,1].set_xlabel("Countries",fontsize=17,fontname='Arial',fontweight='bold')
ax[0,1].set_ylabel("2000 Population (In Billions)",fontsize=17,fontname='Arial',fontweight='bold')
ax[0,1].set_xticks(x["country"])
ax[0,1].set_xticklabels(x["country"],rotation=90,fontsize=15)
ax[0,1].legend(fontsize=17)
ax[0,1].grid(True)
ax[0,1].set_title("Population in 2000",fontweight='bold',fontsize=17)

ax[1,0].bar(x["country"],x["1990 population"],width=0.75,color='purple',label='1990 Population',edgecolor='black')
ax[1,0].set_xlabel("Countries",fontsize=17,fontname='Arial',fontweight='bold')
ax[1,0].set_ylabel("1990 Population (In Billions)",fontsize=17,fontname='Arial',fontweight='bold')
ax[1,0].set_xticks(x["country"])
ax[1,0].set_xticklabels(x["country"],rotation=90,fontsize=15)
ax[1,0].legend(fontsize=17)
ax[1,0].grid(True)
ax[1,0].set_title("Population in 1990",fontweight='bold',fontsize=17)

ax[1,1].bar(x["country"],x["1970 population"],width=0.75,color='olive',label='1970 Population',edgecolor='black')
ax[1,1].set_xlabel("Countries",fontsize=17,fontname='Arial',fontweight='bold')
ax[1,1].set_ylabel("1970 Population (In Billions)",fontsize=17,fontname='Arial',fontweight='bold')
ax[1,1].set_xticks(x["country"])
ax[1,1].legend(fontsize=17)
ax[1,1].grid(True)
ax[1,1].set_xticklabels(x["country"],rotation=90,fontsize=15)

plt.suptitle("Population in Countries Through Different Years",fontsize=29,fontweight='bold',y=0.99)
ax[1,1].set_title("Population in 1970",fontweight='bold',fontsize=17)
plt.tight_layout()
plt.show()
```

# Scatter Graph

```python
#plotting a scatter graph
fig, ax=plt.subplots()
ax.scatter(x["country"],x["2023 population"],color='darkorange',label='2023 population')
ax.scatter(x["country"],x["2022 population"],color='olive',label='2022 population')
ax.scatter(x["country"],x["2020 population"],color='cornflowerblue',label='2020 population')
ax.scatter(x["country"],x["2015 population"],color='brown',label='2015 population')
ax.scatter(x["country"],x["2010 population"],color='aqua',label='2010 population')
ax.scatter(x["country"],x["2000 population"],color='slateblue',label='2000 population')
ax.scatter(x["country"],x["1990 population"],color='mediumorchid',label='1990 population')
ax.scatter(x["country"],x["1980 population"],color='yellow',label='1980 population')
ax.scatter(x["country"],x["1970 population"],color='purple',label='1970 population')
ax.set_xlabel("Country",fontweight='bold')
ax.set_ylabel("Population (In Billions)",fontweight='bold')
ax.set_title("Population Variability Among Countries",fontweight='bold')
fig.legend(loc='upper right',bbox_to_anchor=(1.26,0.940))
ax.set_xticks(x["country"])
ax.grid(True)
ax.set_xticklabels(x["country"],rotation=90)
plt.tight_layout()
plt.show()
```

# Heat Map

```python
#creating a correlation matrix
x_corr=x.corr()
x_corr
```

```python
sel_col=['1970 population','1990 population','2022 population','2023 population']
sel_dat=x[sel_col]
corr_matrix=sel_dat.corr(method='spearman')
plt.title('Correlation Matrix')
print(corr_matrix)

sns.heatmap(corr_matrix,cmap='cividis',annot=True)
plt.show()
```
